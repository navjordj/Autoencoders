```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 

from skimage import io
from skimage.viewer import ImageViewer

import os
```

```python
import tensorflow as tf 
from tensorflow.keras.preprocessing.image import img_to_array, load_img
```

```python
DATA_FOLDER = 'data/img_align_celeba/img_align_celeba'

```

```python
images = os.listdir(DATA_FOLDER)
N_IMAGES = len(images)
N_IMAGES
```

```python
img = io.imread(DATA_FOLDER + '/000002.jpg')
print(img.shape)
#viewer = ImageViewer(img)
#viewer.show()
```

```python
fig, ax = plt.subplots(1,5, figsize=(12, 12)) 

for i, axis in enumerate(ax):
    img = io.imread(DATA_FOLDER + f'/00000{i+1}.jpg')
    axis.set_xticks([])
    axis.set_yticks([])
    axis.set_title(f'/00000{i+1}.jpg')
    axis.imshow(img)
```

```python
def get_data(data_folder):
    X = np.empty(shape=(N_IMAGES, 128, 128, 1))
    for i, filename in enumerate(data_folder):
        img = img_to_array(load_img(DATA_FOLDER + '/' + filename, target_size=(128, 128, 3), color_mode='grayscale'))
        X[i] = img

    X = np.array(X).astype('float32')
    X = X / 255

    return X

data = get_data(images)
```

```python
data.shape
```

```python
fig, ax = plt.subplots(1,5, figsize=(12, 12)) 

for i, axis in enumerate(ax):
    img = data[i]
    axis.set_xticks([])
    axis.set_yticks([])
    axis.set_title(f'/00000{i+1}.jpg')
    axis.imshow(img)
```
